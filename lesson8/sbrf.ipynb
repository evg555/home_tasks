### Создать приложение, которое будет из готового файла с данными «Сбербанка»
### выводить результат по параметрам:- Тип данных- Интервал дат- Область

### Визуализировать выводимые данные с помощью графика

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('opendata.csv')
```

```python
params_list = df['name'].unique()
for i, item in enumerate(params_list):
    print(i+1, item)


param = input("Выберите номер параметра из списка:")
```

```python
region_list = df['region'].unique()
for i, item in enumerate(region_list):
    print(i+1, item)


region = input("Выберите номер региона из списка:")
```

```python
date_start = input("Введите начальную дату в формате YYYY-MM-DD:")
date_end = input("Введите конечную дату в формате YYYY-MM-DD:")
```

```python
df.sort_values(by=['date'])

if not param or not params_list[int(param)]:
    param = 1

if not region or not region_list[int(region)]:
    region = 1 

if not date_start:
    date_start = df.iloc[0]['date'] 
    
if not date_end:
    date_end = df.iloc[-1]['date']   


result = df.loc[(df['name'] == params_list[int(param)-1]) & (df['region'] == region_list[int(region)-1]) 
                & (df['date'] >= date_start) & (df['date'] <= date_end)]
```

```python
plt.figure(figsize=(14,10))
plt.grid()
plt.title(result.iloc[1]['name'],fontsize=17)
plt.xticks(rotation=45)
plt.plot(result['date'],result['value'],'b-')
```
