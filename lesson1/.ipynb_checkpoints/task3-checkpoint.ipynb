### 1. Даны 4 последовательности. Необходимо:
#### a. исследовать их на монотонность; 
#### b. исследовать на ограниченность;
#### c. найти пятый по счету член.


##### 1. Последовательность возрастает, так как 2 ** n всегда будет больше n для n > 1. Последовательность ограниченна только снизу, предел равен 1, сверху - неограниченна. Пятый член равен 2 ** 5 - 5 = 27

##### 2. Последовательность возрастает от -1 до 0-. Последовательность ограниченна снизу, предел равен -1 и сверху, предел стремится к 0. Пятый член равен 1 / (1 - 5) = -0,25

##### 3. Последовательность не монотонна. Последовательность неограниченна ни снизу, ни сверху. Пятый член равен -1 ** 5 + sqr(2*5) ~ -2.7

##### 4. Последовательность убывает. Последовательность ограниченна сверху, предел равен 2 и снизу, предел равен 1. Пятый член равен 1 + 1/5**2 = 1.04

### 2. Найти 12-й член заданной неявно последовательности

#### Каждый последующий член последовательности больше предыдущего на 6. Тогда последовательность можно записать, как 128 + 6*(n - 1). Тогда 12 член будет равен 128 + 6*(12-1) = 128 + 66 = 194
